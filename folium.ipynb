```python
import os
import folium

wa98 = 'data/perZIPgeojson/9/8/'

m = folium.Map(
    location=[47.6, -122.334958],
    zoom_start=10
)

def render_all(start_dir):
    for root, dir, files in os.walk(start_dir):
        for f in files:
            if f.endswith('.json'):
                json_path = os.path.join(root, f)
                #print(json_path)
                folium.GeoJson(
                    json_path,
                    name='geojson'
                ).add_to(m)

render_all(os.path.join(wa98, '1'))

folium.LayerControl().add_to(m)

m
```

```python
import os
import folium

wa98 = 'data/perZIPgeojson/9/8/'

m = folium.Map(
    location=[47.6, -122.334958],
    zoom_start=10
)

def render_all(start_dir):
    for root, dir, files in os.walk(start_dir):
        for f in files:
            if f.endswith('.json'):
                json_path = os.path.join(root, f)
                #print(json_path)
                folium.GeoJson(
                    json_path,
                    name='geojson',
                    smooth_factor=5,
                ).add_to(m)

render_all(os.path.join(wa98, '1'))

folium.LayerControl().add_to(m)

m
```

```python
import os
import folium

from cloudict import WebJSON

m = folium.Map(location=[47.57, -122.23], zoom_start=12)

island = WebJSON('https://raw.githubusercontent.com/yyu/GeoJSON-US/master/perZIPgeojson/9/8/0/98040.json')

folium.GeoJson(island.as_dict(), name='island').add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
import folium
import geopandas as gpd
import pysal as ps

shpfile = '/Users/yy/Downloads/tl_2017_us_zcta510/tl_2017_us_zcta510.shp'
us = gpd.read_file(shpfile)
usa = us.rename(columns={'geometry': 'boundary'}).set_geometry('boundary').to_crs({'init': 'epsg:4326'})
wa981 = usa[(usa.GEOID10 >= '98100') & (usa.GEOID10 < '98200')]
```

```python
m = folium.Map(location=[47.57, -122.23], zoom_start=8)

m.choropleth(geo_data=wa981, data=wa981,
                columns=['GEOID10', 'ALAND10'], key_on='feature.properties.GEOID10',
                fill_color='YlOrRd', fill_opacity=0.5, line_opacity=0.2)

m
```
