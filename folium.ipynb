```python
import os

wa98 = 'data/perZIPgeojson/9/8/'

m = folium.Map(
    location=[47.621795, -122.334958],
    zoom_start=10
)

for zipcode in [98109, 98121, 98040, 98052, 98028]:
    json_path = '{0}{1}/{2}.json'.format(wa98, str(zipcode)[2], zipcode)

    folium.GeoJson(
        json_path,
        name='geojson'
    ).add_to(m)

folium.LayerControl().add_to(m)

m
```
