```python
import os

def files_in_folder(start_dir, is_wanted=lambda full_path: True):
    for root, dirs, files in os.walk(start_dir):
        all_paths = (os.path.join(root, filename) for filename in files)
        wanted_paths = [path for path in all_paths if is_wanted(path)]

        for path in wanted_paths:
            yield path
```

```python
import re

def whitelisted(filename):
    return re.match(r'.*981..\.json', filename)

files_of_interest = list(files_in_folder('data/perZIPgeojson/9/8/', whitelisted))

geojson = {
    "type": "FeatureCollection",
    "features": []
}

for filename in files_of_interest:
    print(filename)
    with open(filename) as f:
        s = f.read()
        j = json.loads(s)
        geojson['features'].append(j)
```

```python
gj = GeoJSON(geojson)
gj
```
